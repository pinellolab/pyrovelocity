```python
import scanpy as sc
import scvelo as sv

adata = sc.read('data/processed/larry_processed.h5ad')
print(adata)
sc.pl.umap(adata, color=['time_info', 'state_info'])
```

```python
time = adata.obs['time_info'].values
time_dic = []
for t in time:
    if t not in time_dic:
        time_dic.append(t)
print('time_info:', time_dic)
adata = adata[adata.obs['time_info'] == 6.0]

print(adata.obs['state_info'].cat.categories)
adata = adata[adata.obs['state_info'] != 'Undifferentiated']

print(adata)
sc.pl.embedding(adata, 'X_emb', color=['time_info', 'state_info']) 

sc.pp.neighbors(adata)
sc.tl.umap(adata)
sc.pl.umap(adata, color=['time_info', 'state_info']) 

```
